Create a new virtual environment using Python 3.12:

Windows (PowerShell or Command Prompt):

```powershell
py -3.12 -m venv venv
```

Activate the environment:

- PowerShell: `venv\Scripts\Activate.ps1` (may need `Set-ExecutionPolicy -Scope Process -ExecutionPolicy Bypass`)
- Generic (works in many shells): `venv\Scripts\activate`
- Command Prompt (cmd.exe): `venv\Scripts\activate.bat`

Common next steps after activation:

```powershell
python -m pip install --upgrade pip
pip install -r requirements.txt
```

Deactivate when finished:

``
deactivate
``

Notes:

- If the `py` launcher isn't available, use `python -m venv venv` instead.
- On Windows PowerShell you may need to allow script execution for the session to run `Activate.ps1`.
